###Chapter 5.1:Series

```python
import pandas as pd 
import numpy as np

prices  = {'CPU': 300, 'GPU': 500, 'RAM': 80, 'SSD': np.nan}
components_prices = pd.Series(prices)
components_prices.name = "Euro_price"
components_prices.index.name = "Component_type"

ron_prices = components_prices*5
print(ron_prices.isna())

ron_prices[ron_prices>400]
```
