#Chapter 5.1:Index Objects  
###Context: You manage databases for two technical warehouses stocking PC components. You want to use Pandas Index structures to identify shared and unique components, and to learn how to safely manipulate these labels while adhering to immutability principles.

```python
import numpy as np
import pandas as pd

index_deposit_A = pd.Index(['CPU', 'GPU', 'RAM', 'SSD', 'Motherboard'])
index_deposit_B = pd.Index(['RAM', 'SSD', 'HDD', 'Cooler', 'PSU'])

common_components = index_deposit_A.intersection(index_deposit_B)
all_components = index_deposit_A.union(index_deposit_B)
first_three = index_deposit_A[:3]
print('Total size:',np.size(all_components))

index_deposit_A_modified = pd.Index(['Processor']).append(index_deposit_A[1:])
print(index_deposit_A_modified)
```
