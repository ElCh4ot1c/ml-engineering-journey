#Chapter 5.2:Reindexing  
###Context: You are managing an IoT temperature monitoring system in a pharmaceutical warehouse where medicines must be kept at strict temperatures. Your sensor transmits data every two hours, but the central reporting system requires a complete hourly report. You will need to use reindexing and a filling technique to avoid leaving gaps (NaN) in the report.

```python
import numpy as np
import pandas as pd

senzor_temps = pd.Series([18.2, 18.5, 19.0, 18.8], index=[0, 2, 4, 6])
hour_report = senzor_temps.reindex(np.arange(7))
print("Hour report between 1-6 hour:\n",hour_report)
hour_report_updated = senzor_temps.reindex(np.arange(7), method="ffill")
print("Updated hour report between 1-6 hour:\n",hour_report_updated)

df_deposit = pd.DataFrame({'Temp': [18.2, 19.0],'Humidity': [45, 48]}, index=['Zone_A', 'Zone_C'])
df_deposit = df_deposit.reindex(index=['Zone_A', 'Zone_B', 'Zone_C'],fill_value=0)
print("Final report:\n",df_deposit)
```

#Chapter 5.2:Dropping Entries from an Axis  
###Context: You work on the cybersecurity team of a tech company. A group of hackers has attacked the cloud infrastructure. You have extracted a DataFrame showing the status of the servers, but the investigation revealed that the hackers installed malware on "Server_Beta" and completely corrupted the data in the "Uptime_Days" column.  
  
Your task is to use the `.drop()` function to remove the compromised entries and unnecessary columns before submitting the clean report to management.

```python
import pandas as pd
import numpy as np

attack_data = {
    'IP_Address': ['192.168.1.10', '10.0.0.5', '172.16.0.22'],
    'Traffic_GB': [150.5, 2900.2, 45.1], 
    'Uptime_Days': [42, np.nan, 115]
}
df_security = pd.DataFrame(attack_data, index=['Server_Alpha', 'Server_Beta', 'Server_Gamma'])
df_isolated = df_security.drop('Server_Beta')
df_final_report = df_isolated.drop(columns='Uptime_Days')
print(df_final_report)
```

#Chapter 5.2:Indexing, Selection, and Filtering  
###Context: You are a data analyst at a video game company. You have a table containing statistics for three famous players. The server has logged their matches, and you need to quickly extract specific data points for the marketing department using .loc, .iloc, and logical filters.

```python
import pandas as pd
import numpy as np

players_data = {
    'Score': [2500, 4100, 1200],
    'Hours_played': [120, 350, 40],
    'Premiu': [True, True, False]
}
df_gaming = pd.DataFrame(players_data, index=['Player_X', 'Player_Y', 'Player_Z'])
print(df_gaming.iloc[0:2])
print(df_gaming.loc[:,['Score','Hours_played']])
print(df_gaming[df_gaming['Hours_played']>100])
```

#Chapter 5.2:Function Application and Mapping  
##Context:You are building a script to evaluate if your Idleon characters are ready for a Boss Fight. To do this, you need to calculate their final combat rating and dynamically assign a tier, handling conditions on both columns and rows.


```python
import numpy as np
import pandas as pd



# Base combat stats for characters
party_stats = pd.DataFrame({
    'Attack': [210, 90, 350, 120],
    'Speed': [45, 120, 30, 80],
    'Defense': [80, 40, 150, 65]
}, index=['Warrior_1', 'Archer_1', 'Squire_2', 'Mage_1'])
party_stats=party_stats.apply(lambda value: value - value.min(), axis=0)
party_stats['Combat_Rating'] = party_stats.apply(lambda col: col['Attack']*3+col['Speed']*1.5+col['Defense'],axis=1)

party_stats['Tier'] = np.where(party_stats['Combat_Rating']>500, 'Boss_Ready', 'Needs_Gear')
print(party_stats)
```

#Chapter 5.2:Sorting and Ranking

```python
import pandas as pd
import numpy as np

# Raid performance logs
raid_results = pd.DataFrame({
    'Damage_Dealt': [450000, 120000, 450000, 950000, 310000],
    'Deaths': [1, 4, 0, 2, 1]
}, index=['Warrior_A', 'Mage_A', 'Warrior_B', 'Archer_A', 'Mage_B'])

sorted_raid = raid_results.sort_values(['Damage_Dealt','Deaths'],ascending=[False,True])
print(sorted_raid)

sorted_raid["Damage_Rank"] = sorted_raid['Damage_Dealt'].rank(ascending=False)
print(sorted_raid)

sorted_raid["Strict_Rank"] = sorted_raid['Damage_Dealt'].rank(ascending=False, method="min")
print(sorted_raid)
```

#Chapter 5.2:Final Test  
###Context:You are managing an MMORPG database at the end of a competitive season. You need to merge player data from two different regional servers, clean up formatting errors, filter out low-level testers, apply a seasonal balance patch formula, and generate the global prize leaderboard.

```python
import pandas as pd
import numpy as np

# SERVER 1: Europe (Uses uppercase class names as index, contains some missing data)
server_EU = pd.DataFrame({
    'Level': [45, 115, 88, 12],
    'Gold': [12000, 450000, 180000, 500],
    'Base_DPS': [450.0, 1200.0, 950.0, 110.0]
}, index=['WARRIOR', 'ARCHER', 'MAGE', 'WARRIOR'])

# SERVER 2: North America (Uses lowercase class names as index, has different metrics)
server_NA = pd.DataFrame({
    'Level': [95, 110, 30],
    'Gold': [210000, 500000, 4000],
    'Bonus_Multiplier': [1.15, 1.25, 1.05]
}, index=['mage', 'archer', 'warrior'])
server_EU.index=server_EU.index.str.lower()
print(server_EU)

server_EU = server_EU[server_EU['Level']>40]
server_NA = server_NA[server_NA['Level']>40]

server_EU.index = server_EU.index + '_EU'
server_NA.index = server_NA.index + '_NA'

global_database = server_NA.add(server_EU, fill_value=0)
global_database = global_database.fillna(0)

print(global_database.index.is_unique)
global_database = global_database.reset_index(names='Class')
print(global_database)

global_database['Final_Score'] = np.where(global_database['Bonus_Multiplier']==0 , global_database['Base_DPS'] * 1.5 + global_database['Gold'] * 0.01 , np.where(global_database['Base_DPS'] == 0, global_database['Level'] * 10 * global_database['Bonus_Multiplier'] + global_database['Gold'] * 0.01,0) )
global_database=global_database.sort_values(by='Final_Score',ascending=False)
global_database['Global_Rank'] = global_database["Final_Score"].rank(ascending=False,method="min")
print("Top 3:\n",global_database.iloc[:3,:])
```
