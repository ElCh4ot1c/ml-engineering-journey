#Chapter 5.3:  Summarizing and Computing Descriptive Statistics
##Context:You are managing the database for a military strategy outpost. Four commanders have been purchasing gold rations over three months to prepare for a major siege. Due to sporadic network connectivity from the front lines, some transaction data was lost, resulting in missing values (NaN).

```python
import pandas as pd
import numpy as np

# Monthly expenditures in Gold Rations
shop_data = pd.DataFrame({
    'January': [5000, 2200, np.nan, 4100],
    'February': [3100, np.nan, 1500, 6000],
    'March': [4200, 1800, 2900, np.nan]
}, index=['Commander_Alpha', 'Commander_Bravo', 'Commander_Charlie', 'Commander_Delta'])

print('Monthly expenses:\n', shop_data.sum(axis=0))
shop_data['Total_Spent'] = shop_data.sum(axis=1)
print("January average expediture: \n", shop_data["January"].mean(skipna=False))

print("Biggest Spender:\n",shop_data['Total_Spent'].idxmax(axis=0))

print('February complete descriptive statistics summary report: \n',shop_data['February'].describe())
```
