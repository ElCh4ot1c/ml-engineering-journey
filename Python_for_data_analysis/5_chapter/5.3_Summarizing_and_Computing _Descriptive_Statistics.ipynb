#Chapter 5.3:  Summarizing and Computing Descriptive Statistics  
##Context:You are managing the database for a military strategy outpost. Four commanders have been purchasing gold rations over three months to prepare for a major siege. Due to sporadic network connectivity from the front lines, some transaction data was lost, resulting in missing values (NaN).

```python
import pandas as pd
import numpy as np

# Monthly expenditures in Gold Rations
shop_data = pd.DataFrame({
    'January': [5000, 2200, np.nan, 4100],
    'February': [3100, np.nan, 1500, 6000],
    'March': [4200, 1800, 2900, np.nan]
}, index=['Commander_Alpha', 'Commander_Bravo', 'Commander_Charlie', 'Commander_Delta'])

print('Monthly expenses:\n', shop_data.sum(axis=0))
shop_data['Total_Spent'] = shop_data.sum(axis=1)
print("January average expediture: \n", shop_data["January"].mean(skipna=False))

print("Biggest Spender:\n",shop_data['Total_Spent'].idxmax(axis=0))

print('February complete descriptive statistics summary report: \n',shop_data['February'].describe())
```

#Chapter 5.3:Correlation and Covariance  
##You work as a Data Analyst for an international online store. The security team suspects that certain bots or compromised accounts are launching "credential stuffing" attacks or attempting to manipulate inventory.

You have extracted a sample of statistics for five users to see how their actions correlate (hours spent on the site, number of pages visited, orders placed, and failed login attempts).

```python
import pandas as pd
import numpy as np

# User behavior logs
traffic_data = pd.DataFrame({
    'Hours_Online': [1.2, 24.5, 3.8, 28.0, 0.5],
    'Pages_Visited': [15, 890, 45, 915, 8],
    'Failed_Logins': [0, 42, 1, 55, 0],
    'Orders_Placed': [1, 0, 2, 0, 0]
}, index=['User_Alpha', 'User_Beta', 'User_Gamma', 'User_Delta', 'User_Epsilon'])

print('Traffic datas covariance:\n',traffic_data.cov())
print('Traffic datas correlation:\n',traffic_data.corr())

print(traffic_data['Pages_Visited'].corr(traffic_data['Failed_Logins']))
```

#Chapter 5.3: Unique Values, Value Counts, and Membership  
##You are working as a Junior ML Engineer for a cybersecurity company. The threat intelligence team has provided you with a raw dataset containing security logs from various cloud servers. Before feeding this data into an Anomaly Detection Model, you need to explore the categorical variables, check for specific attack patterns, and find out how the labels are distributed.

```python
import pandas as pd
import numpy as np

# Raw security logs dataset
security_logs = pd.DataFrame({
    'Alert_Type': ['Brute_Force', 'Malware', 'Brute_Force', 'DDOS', 'Malware', 'Brute_Force', 'Phishing'],
    'Severity': ['High', 'Critical', 'High', 'Critical', 'Low', 'High', 'Medium'],
    'Error_Code': [401, 500, 401, 503, 500, 401, 404]
})

print("Unique types of alerts:\n", security_logs['Alert_Type'].unique())
print("Alerts appearences:\n", security_logs['Alert_Type'].value_counts())

print('Filtered DataFrame:\n', security_logs[security_logs['Alert_Type'].isin(['Brute_Force','DDOS'])])

print('2 columns pattern:\n', security_logs.value_counts(['Alert_Type','Severity']))
```
