#Chapter 6.1:Reading and Writing Data in Text Format  
##Context: A marketing team sent you a messy customer behavior file. It uses an unusual separator, contains a useless introduction line, and uses the text string "UNKNOWN_DATA" whenever a value is missing. You need to load it into a clean DataFrame for an upcoming Customer Churn ML model.

```python
import pandas as pd
import io

# Simulated raw text file on disk
raw_marketing_data = """# Internal Marketing Export - Date: 2026-07-12
CustomerID|Age|Annual_Income|Churn_Label
1001|34|550000|No
1002|UNKNOWN_DATA|820000|Yes
1003|45|UNKNOWN_DATA|No
1004|22|310000|Yes"""

df_data = pd.read_csv(io.StringIO(raw_marketing_data),comment='#', na_values='UNKNOWN_DATA',sep='|')
print('Structured data:\n', df_data)
```

#Chapter 6.1:Reading Text Files in Pieces
##Context: You are working with a huge server transactions log file that is 15GB in size. Loading it all at once will crash your system. You need to prototype a pipeline that loads only a small sample to inspect it, and then processes the whole file in controlled batches.

```python
import pandas as pd
import io

# Simulated huge file containing millions of rows
raw_huge_logs = """Timestamp,Server_ID,Response_Time,Status
10:00:01,SRV_01,0.12,200
10:00:02,SRV_02,0.85,500
10:00:03,SRV_01,0.05,200
10:00:04,SRV_03,2.34,404
10:00:05,SRV_02,0.11,200
10:00:06,SRV_01,0.44,200"""

df_preview = pd.read_csv(io.StringIO(raw_huge_logs), nrows=3)
print("Data preview:\n", df_preview)

chunks = pd.read_csv(io.StringIO(raw_huge_logs), chunksize=2)
for piece in chunks:
    print(piece)
```

#Chapter 6.1: Writing Data to Text Format  
##Context: You have successfully cleaned a dirty dataset, handled missing values, and calculated a new feature called High_Risk. Now you must save this dataset to the hard disk so the ML modeling team can pick it up.

```python
import pandas as pd
import numpy as np
import io

# Cleaned DataFrame ready for production storage
cleaned_ml_data = pd.DataFrame({
    'Feature_A': [1.5, np.nan, 3.8],
    'Feature_B': [10, 20, 30],
    'High_Risk': [True, False, True]
})

cleaned_ml_data.to_csv('cleaned_data.csv',index=False,na_rep='MISSING')
```
