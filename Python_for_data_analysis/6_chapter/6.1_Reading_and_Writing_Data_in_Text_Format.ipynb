#Chapter 6.1:Reading and Writing Data in Text Format  
##Context: A marketing team sent you a messy customer behavior file. It uses an unusual separator, contains a useless introduction line, and uses the text string "UNKNOWN_DATA" whenever a value is missing. You need to load it into a clean DataFrame for an upcoming Customer Churn ML model.

```python
import pandas as pd
import io

# Simulated raw text file on disk
raw_marketing_data = """# Internal Marketing Export - Date: 2026-07-12
CustomerID|Age|Annual_Income|Churn_Label
1001|34|550000|No
1002|UNKNOWN_DATA|820000|Yes
1003|45|UNKNOWN_DATA|No
1004|22|310000|Yes"""

df_data = pd.read_csv(io.StringIO(raw_marketing_data),comment='#', na_values='UNKNOWN_DATA',sep='|')
print('Structured data:\n', df_data)
```
