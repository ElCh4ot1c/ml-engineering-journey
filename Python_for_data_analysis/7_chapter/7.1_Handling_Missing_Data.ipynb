##Context:You are working solo as a Data Engineer for a Smart City project in Chișinău. You have deployed multiple IoT sensors across the city to monitor traffic density and air quality. However, due to unstable Wi-Fi connections, sensor battery depletion, and hardware glitches, the raw data collected over the last 24 hours is messy and contains numerous missing values (NaN).

Before this data can be fed into an AI model that predicts traffic congestion and pollution spikes, you must build a robust data-cleaning pipeline using Pandas to handle these missing values scientifically.

```python
import pandas as pd
import numpy as np

# Setting up the mock corrupted dataset
data = {
    'Timestamp': pd.date_range(start='2026-07-13 00:00', periods=10, freq='h'),
    'Sensor_ID': ['SNS_01', 'SNS_02', 'SNS_01', 'SNS_03', 'SNS_02', 'SNS_01', 'SNS_03', 'SNS_02', 'SNS_01', 'SNS_03'],
    'Traffic_Count': [120, np.nan, 85, 210, np.nan, np.nan, 195, 110, 95, np.nan],
    'PM2_5': [12.5, 45.2, np.nan, 18.1, 41.8, np.nan, np.nan, 38.5, 14.2, 22.0],
    'Sensor_Status': ['Active', 'Active', np.nan, 'Maintenance', 'Active', np.nan, 'Active', np.nan, 'Active', 'Active']
}


df = pd.DataFrame(data)

print('Total number of missing values per column:\n', df.isna().sum(),'\n','The percentage of missing values:\n',df.isna().mean()*100)

# Injecting a completely empty row (except keys) to test Task 2
df.iloc[5, 2:] = np.nan  # Row 5 becomes completely empty for data columns
df = df.dropna(subset=['Traffic_Count'])
print('Available rows:\n', df)
df['Sensor_Status'] = df['Sensor_Status'].fillna('Unknown')
df['PM2_5'] = df['PM2_5'].fillna(df.groupby('Sensor_ID')['PM2_5'].transform('median'))
print(df)
print("-" * 30)
#df.isna()
```
