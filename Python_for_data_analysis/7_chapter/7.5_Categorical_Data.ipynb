#Chapter 7.5:Categorical Data  
##You are a Data Analyst at a telecommunications company. You have received a raw dataset containing 1,000 customers (churn_data). The current dataset uses unoptimized string objects for plan types, support tiers, and consumption categories, wasting unnecessary memory and preventing logical sorting (for instance, "Basic" does not naturally sort before "Premium").

Your task is to transform and optimize this dataset using only Pandas Categorical features (pd.Categorical, .astype('category'), the pd.cut function, and the .cat accessor).

```python
import pandas as pd
import numpy as np
from pandas.api.types import CategoricalDtype

np.random.seed(42)
n_rows = 1000

plans = np.random.choice(["Basic", "Standard", "Premium", "Enterprise"], size=n_rows)
regions = np.random.choice(["North", "South", "East", "West", "Central"], size=n_rows)
support_tier = np.random.choice(["Tier 1", "Tier 2", "Tier 3"], size=n_rows)
monthly_charges = np.random.uniform(10.0, 150.0, size=n_rows)
contract_status = np.random.choice(["Active", "Paused", "Cancelled", "Expired"], size=n_rows)

df = pd.DataFrame({
    "customer_id": range(5001, 5001 + n_rows),
    "region": regions,
    "plan_type": plans,
    "support_tier": support_tier,
    "monthly_charges": monthly_charges,
    "status": contract_status
})

df['region'] = df['region'].astype("category")

ordered_cat = CategoricalDtype(categories=["Basic" , "Standard" , "Premium" , "Enterprise"],ordered=True)
df['plan_type'] = df['plan_type'].astype(ordered_cat)

ordered_cat2 = CategoricalDtype(categories=["Tier 1" , "Tier 2" , "Tier 3"],ordered=True)
df['support_tier'] = df['support_tier'].astype(ordered_cat2)
charge_groups = ["Low", "Medium", "High"]
df['charge_group'] = pd.cut(df["monthly_charges"],bins=3,labels=charge_groups)

df["plan_type"] = df['plan_type'].cat.add_categories(['Custom-VIP'])
df["region"] = df["region"].cat.rename_categories({
    'North': 'Nord',
    'South': 'Sud',
    'West' : 'Vest',
    'East' : 'Est'
})

df['status'] = df['status'].astype('category')  
df = df[df['status'] == 'Active']
df['status'] = df['status'].cat.remove_unused_categories()
df['plan_code'] = df['plan_type'].cat.codes
df
```
