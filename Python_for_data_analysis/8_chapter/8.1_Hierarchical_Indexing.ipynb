#Chapter 8.1:Hierarchical Indexing   
##You are analyzing a regional retail network. You have raw sales and inventory data spread across multiple stores, product categories, and quarters. To build an efficient analytical pipeline, you need to structure the dataset using Hierarchical Indexing (MultiIndex) to quickly slice, reshuffle, and aggregate metrics across different dimensions without flattening the data into basic columns.

Your task is to structure and manipulate this multi-index dataset using only MultiIndex concepts (pd.MultiIndex, .loc indexing with tuples/IndexSlice, swaplevel, sort_index, unstack/stack, and level-based aggregations).

```python
import pandas as pd
import numpy as np

np.random.seed(42)

# --- 1. Definirea Datelor ---
tuples = [
    ('Chisinau', 'Store_1', 'Electronics'),
    ('Chisinau', 'Store_1', 'Clothing'),
    ('Chisinau', 'Store_2', 'Electronics'),
    ('Chisinau', 'Store_2', 'Clothing'),
    ('Balti', 'Store_1', 'Electronics'),
    ('Balti', 'Store_1', 'Clothing'),
    ('Balti', 'Store_2', 'Electronics'),
    ('Balti', 'Store_2', 'Clothing')
]

multi_idx = pd.MultiIndex.from_tuples(tuples, names=['city', 'store', 'category'])

data = np.random.randint(100, 1000, size=(8, 4))
columns = pd.MultiIndex.from_product(
    [['Q1', 'Q2'], ['Revenue', 'Units_Sold']],
    names=['quarter', 'metric']
)

df = pd.DataFrame(data, index=multi_idx, columns=columns)
idx = pd.IndexSlice

# --- TASK 1: Selection & Slicing ---
# 1a. Electronics în Chisinau
res_1a = df.loc[idx['Chisinau', :, 'Electronics'], idx[['Q1', 'Q2'], :]]

# 1b. Q2 Revenue pentru Store_1
res_1b = df.loc[idx[:, 'Store_1', :], idx['Q2', 'Revenue']]

# --- TASK 2: Reordering & Sorting ---
df_reordered = df.swaplevel('category', 'city').sort_index(axis=0)

# --- TASK 3: Level-Based Aggregation ---
df_grouped = df_reordered.groupby(level='category').sum()

# --- TASK 4: Reshaping (Unstack & Stack) ---
df_reshaped = df_reordered.unstack(level='category').stack(level='quarter')
```
